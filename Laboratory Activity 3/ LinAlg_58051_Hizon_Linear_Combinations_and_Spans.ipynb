```
import numpy as np
import matplotlib.pyplot as plt
```

$$A = \begin{bmatrix} 1\\6 \\\end{bmatrix} , B = \begin{bmatrix} 5\\8 \\\end{bmatrix} $$

```
A = np.array([1, 6])
B = np.array([5, 8])
R_AB = np.add(A, B)

plt.xlim(-15, 15)
plt.ylim(-10, 25)

plt.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale = 1, color= "goldenrod", label='A')
plt.quiver(A[0], A[1], B[0], B[1], angles='xy', scale_units='xy', scale=1, color='cadetblue', label='B')
plt.quiver(0, 0, R_AB[0], R_AB[1], angles='xy', scale_units='xy', scale=1, color ='seagreen', label="Resultant Vector")

plt.legend()
plt.grid()
plt.show()
```

$$C = \begin{bmatrix} 3\\8 \\\end{bmatrix} , D = \begin{bmatrix} 2\\1 \\\end{bmatrix} $$

```
C = np.array([3, 8])
D = np.array([2, 1])
R_CD = np.add(C, D)

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.quiver(0, 0, C[0], C[1], angles='xy', scale_units='xy', scale = 1, color= "limegreen", label='C')
plt.quiver(C[0], C[1], D[0], D[1], angles='xy', scale_units='xy', scale=1, color='salmon', label='D')
plt.quiver(0, 0, R_CD[0], R_CD[1], angles='xy', scale_units='xy', scale=1, color='purple',label='Resultant Vector')

plt.legend()
plt.grid()
plt.show()
```

$$E = \begin{bmatrix} 6\\-1 \\\end{bmatrix} , F = \begin{bmatrix} 0\\-3 \\\end{bmatrix} $$

```
E = np.array([6, -1])
F = np.array([0, -3])
R_EF = np.add(E, F)

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.quiver(0, 0, E[0], E[1], angles='xy', scale_units='xy', scale = 1, color= "slategrey", label='E')
plt.quiver(E[0], E[1], F[0], F[1], angles='xy', scale_units='xy', scale=1, color='crimson', label='F')
plt.quiver(0, 0, R_EF[0], R_EF[1], angles='xy', scale_units='xy', scale=1, color='darkblue', label='Resultant Vector')

plt.legend()
plt.grid()
plt.show()
```
