# **PART 1: Monthly Sales**

```python
import numpy as np
```

```python
## Vectorizing the data
sales_mat = np.array(df_sales.set_index('flavor'))
prices_mat = np.array(df_prices)[0]
costs_mat = np.array(df_prices)[1]

prices_mat = prices_mat.reshape(12, 1) # Reshapping prices_mat array
costs_mat = costs_mat.reshape(12, 1) # Reshapping costs_mat array

## Printing the shapes of sales_mat, prices_mat, and costs_mat
print(sales_mat.shape)
print(prices_mat.shape)
print(costs_mat.shape)
```

```python
# A function that returns and prints the monthly sales and profit of each month.
def monthly_sales(price, cost, sales):
    monthly_revenue = sum(np.multiply(sales, price))
    monthly_costs = sum(np.multiply(sales, cost))
    monthly_profits = np.subtract(monthly_revenue, monthly_costs)  
    return monthly_revenue.flatten(), monthly_costs.flatten(), monthly_profits.flatten()
```

```python
month_rev, month_cost, month_profit = monthly_sales(prices_mat, costs_mat, sales_mat)

#Printing the monthly and yearly values of revenue, cost, and profit.
print("REVENUE")
print("Monthly Revenue: \n", month_rev)
print("Yearly Revenue: ", sum(month_rev), "\n")

print("COST")
print("Monthly Cost: \n", month_cost)
print("Yearly Cost: ", sum(month_cost), "\n")

print("PROFIT")
print("Monthly Profit: \n", month_profit)
print("Yearly Profit: ", sum(month_profit), "\n")
```

# **PART 2: Flavor Sales**

```python
# A function that returns and prints the flavor profits of the whole year. 
def flavor_sales(price, cost, sales):
    flavor_revenue = np.multiply(sales, price)
    flavor_costs = np.multiply(sales, cost)
    flavor_profits = np.subtract(flavor_revenue, flavor_costs)  
    return flavor_profits.flatten()
```

```python
flavor_profit = flavor_sales(prices_mat, costs_mat, sales_mat)

## Values of the profit for each flavor
flavor_1 = []
flavor_2 = []
flavor_3 = []
flavor_4 = []
flavor_5 = []
flavor_6 = []
flavor_7 = []
flavor_8 = []
flavor_9 = []
flavor_10 = []
flavor_11 = []
flavor_12 = []

# A loop that adds the profit value to their corresponding variables above (eg. flavor_1, etc.)
for iteration in flavor_profit:
  if len(flavor_1) <= 11:
    flavor_1.append(iteration)
  elif len(flavor_2) <= 11:
    flavor_2.append(iteration)
  elif len(flavor_3) <= 11:
    flavor_3.append(iteration)
  elif len(flavor_4) <= 11:
    flavor_4.append(iteration)
  elif len(flavor_5) <= 11:
    flavor_5.append(iteration)
  elif len(flavor_6) <= 11:
    flavor_6.append(iteration)
  elif len(flavor_7) <= 11:
    flavor_7.append(iteration)
  elif len(flavor_8) <= 11:
    flavor_8.append(iteration)
  elif len(flavor_9) <= 11:
    flavor_9.append(iteration)
  elif len(flavor_10) <= 11:
    flavor_10.append(iteration)
  elif len(flavor_11) <= 11:
    flavor_11.append(iteration)
  elif len(flavor_12) <= 11:
    flavor_12.append(iteration)

# Getting the profit of each flavor for the whole year.
profit_Flavor = np.array([sum(flavor_1), sum(flavor_2), sum(flavor_3), sum(flavor_4), sum(flavor_5), sum(flavor_6), sum(flavor_7), sum(flavor_8),
                          sum(flavor_9), sum(flavor_10), sum(flavor_11), sum(flavor_12)])

# Printing the monthly and yearlt profit of each flavor
# Refer to the arrangement of flavors provided on Bebang's Sales
print("Legend (Flavors): ")
print("\t 1. RV - Red Velvet \n \t 2. O  - Oreo \n \t 3. SG - Sugar Glazed \n \t 4. AH - Almond Honey \n \t 5. M  - Matcha \n \t 6. SC - Strawberry Cream")
print("\t 7. BSB- Brown Sugar Boba \n \t 8. F&N- Fruits and Nuts \n \t 9. DC - Dark Chocolate \n \t10. C&O- Chocolate and Orange \n \t11. CM - Choco Mint")
print("\t12. CBN- Choco Butter Naught \n")
print("Monthly Profit of Each Flavor: \n")
print("  Jan    Feb   Mar   Apr  May \t June  July  Aug   Sept  Oct  Nov   Dec")
print(flavor_profit, "\n")

print("Yearly Profit of Each Flavor?: \n") 
print("  RV     O    SG    AH     M    SC    BSB   F&N    DC   C&O   CM    CBN")
print(profit_Flavor)
```
